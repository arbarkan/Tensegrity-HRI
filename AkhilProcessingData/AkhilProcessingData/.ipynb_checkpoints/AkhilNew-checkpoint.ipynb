```python
#Force Divider Function
def force_divider(x):
    k_outer = 3.738
    k_inner = 0.368
    dist = x/k_inner
    return (k_outer + k_inner)*dist

#FSR model function
def FSR_model(x):
    return 131.2*np.exp(0.7801*x)/1000

#this code converts all raw fsr data in the data matrix to Newtons 
for m in data:
    m[:,1:13] = force_divider(FSR_model(m[:,1:13] *5/1023))  

#d = pd.DataFrame(data[1])
#d.head()
```

```python
#feature function block 
def impulse(data_matrix):
    #total impulse function for a single matrix of data 
    #iterates over data and computes the total impulse for all sensors over time 
    fsr_data = data_matrix[:,1:13]
    impulse = 0
    for col in np.arange(1,13):
        single_FSR = data_matrix[:,col]
        time = data_matrix[:,0]
        for i in np.arange(0,len(time)-1):
            delta_t = time[i+1]- time[i]
            averagef = (single_FSR[i+1] + single_FSR[i])/2 
            impulse = impulse + averagef*delta_t
    return impulse 


def jerk(data_matrix): 
    #iterates over data and computes the max jerk for each sensor over time
    #takes the max jerk over all FSR sensors and returns  
    fsr_data = data_matrix[:,1:13]
    jerk_array = []
    for col in np.arange(1,13):
        single_FSR = data_matrix[:,col]
        time = data_matrix[:,0]
        all_jerks = []
        for i in np.arange(0,len(time)-1):
            delta_t = time[i+1] - time[i]
            delta_f = (single_FSR[i+1] - single_FSR[i])
            all_jerks.append(delta_f/delta_t)
        jerk_array.append(np.amax(all_jerks))
    max_jerk = np.amax(jerk_array)
    return max_jerk

def max_force(data_matrix):
    #max force readings across all sensors for a single matrix of data 
    fsr_data = data_matrix[:,1:13]
    force_array = [] 
    for col in np.arange(1,13):
        single_FSR = data_matrix[:,col]
        max_force_sensor = np.amax(single_FSR)
        force_array.append(max_force_sensor)
    max_force = np.amax(force_array)
    return max_force

def oscillation_freq_old(data_matrix):
    fsr_data = data_matrix[:,1:13]
    length = np.shape(data_matrix)[0] - 1
    total_time = data_matrix[length,0] - data_matrix[0,0]
    freq_array = []
    for col in np.arange(1,13):
        single_FSR = data_matrix[:,col]
        delta_f = 0
        last_delta_f = 0
        total_changes = 0
        for i in np.arange(0,length):
            delta_f = (single_FSR[i+1] - single_FSR[i])
            if delta_f * last_delta_f < 0:
                total_changes+=1 
            last_delta_f = delta_f
        freq_array.append(total_changes/total_time)  
    max_oscillation_freq = np.amax(freq_array)
    return max_oscillation_freq

def oscillation_freq(data_matrix):
    fsr_data = data_matrix[:,1:13]
    length = np.shape(data_matrix)[0] - 1
    total_time = data_matrix[length,0] - data_matrix[0,0]
    freq_array = []
    for col in np.arange(1,13):
        single_FSR = data_matrix[:,col]
        #avg = np.mean(single_FSR)
        avg = 5
        total_passes = 0
        for i in np.arange(0,length):
            current_val = single_FSR[i]
            next_val = single_FSR[i+1]
            if ((current_val > avg) and (next_val < avg)) or ((current_val < avg) and (next_val > avg)):
                total_passes+=1
        freq_array.append(total_passes/total_time)  
    avg_oscillation_freq = np.mean(freq_array)
    return avg_oscillation_freq

def oscillation_freq_new(data_matrix):
    fsr_data = data_matrix[:,1:13]
    length = np.shape(data_matrix)[0] - 1
    peak_frequencies = []
    for col in np.arange(1,13):
        single_FSR = data_matrix[:,col]
        time = data_matrix[:,0]
        function = interp1d(time, single_FSR)
        new_t = np.linspace(data_matrix[0,0],data_matrix[length,0],1000)
        new_data = function(new_t)
    #plt.plot(time, single_FSR)
    #plt.plot(new_t, new_data)
    
    
        time_step = new_t[1] - new_t[0]
        sig_fft = fftpack.fft(new_data)
        power = np.abs(sig_fft)
        sample_freq = fftpack.fftfreq(new_data.size, d=time_step)
        pos_mask = np.where(sample_freq > 0)
        freqs = sample_freq[pos_mask]
        peak_freq = freqs[power[pos_mask].argmax()]
        peak_frequencies.append(peak_freq)
    return np.mean(peak_frequencies)
    
    
    
    
    #fourier = np.fft.fft(new_data)
    #freqs = np.fft.fftfreq(len(fourier))
    #print(freqs.min(), freqs.max())
    #frate = new_t[1] - new_t[0]
    #idx = np.argsort(np.abs(fourier))[1]
    #print(idx)
    #freq = freqs[idx]
    #freq_in_hertz = abs(freq * frate)
    #return freq_in_hertz
    
    
    
    
    #N = new_data.size
    #T = new_t[1] - new_t[0]
    #f = np.linspace(0, 1/T, N)
    #plt.ylabel("Amplitude")
    #plt.xlabel("Frequency [Hz]")
    #plt.bar(f[:N // 2], np.abs(fft)[:N // 2] * 1 / N, width=1.5)  # 1 / N is a normalization factor
    #plt.show()
        
```

```python
#using just nothing, impact, and squeeze data
cropped_data = smooth_data[14:57]
cropped_labels = labels[14:57]

feature_data = []
for m in cropped_data:
    maxf = max_force(m)
    j = jerk(m)
    i = impulse(m)
    features = [maxf, j, i]
    feature_data.append(features)
    
print(np.shape(feature_data))
print(" ")

[X_train, X_test, y_train, y_test] = train_test_split(feature_data, cropped_labels, random_state=0)


#Naive Bayes Implementation 
nb = GaussianNB().fit(X_train, y_train)
print(nb.score(X_train, y_train))
print(nb.score(X_test, y_test))
print("Predicted labels = " + str(nb.predict(X_test)))
print("Correct labels =   " + str(y_test))

print(" ")
#Random Forest Implementation 
rf = RandomForestClassifier().fit(X_train, y_train)
print(rf.score(X_train, y_train))
print(rf.score(X_test, y_test))
print("Predicted labels = " + str(rf.predict(X_test)))
print("Correct labels =   " + str(y_test))
```

```python

def color_bar(trained_model, full_data, window_size = 1000, shift = 1):
    #this function creates a graph of the full data with a color bar which shows the models predictions as the window is shifted
    #red dot = impact prediction 
    #blue dot = nothing prediction
    #green dot = squeeze prediction 
    start_window = 0
    end_window = window_size
    length = (np.shape(full_data))[0]
    impact = 0
    squeeze = 0
    nothing = 0
    while end_window <= length:
        sample = full_data[start_window:end_window,:]
        for c in np.arange(1,13):
            sample[:,c] = signal.savgol_filter(sample[:,c], 21, 3) #should be filtered the same way as original data the model was trained on 
        t = (sample[window_size-1,0]- sample[0,0])/2 + sample[0,0]
        #calculate features 
        maxf = max_force(sample)
        j = jerk(sample)
        i = impulse(sample)
        features = [maxf, j, i]
        prediction = rf.predict([features])[0]
        dot_size = 7 #size of dots on scatter plot 
        if prediction == 2:
            plt.scatter(t, -1, s = dot_size, c ='red')
            impact+=1
        elif prediction == 3:
            plt.scatter(t, -1, s = dot_size, c ='blue')
            nothing+=1
        elif prediction == 4: 
            plt.scatter(t, -1, s = dot_size, c ='green')
            squeeze+=1
        #increase shift to reduce computation time. should be greater than or equal to 1
        start_window+=shift
        end_window+=shift

    plt.plot(full_data[:,0], full_data[:,1:13])
    print('Impact Predictions: ' + str(impact))
    print('Squeeze Predictions: ' + str(squeeze))
    print('Nothing Predictions: ' + str(nothing))


```
