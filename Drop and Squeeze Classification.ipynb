## Preprocessing Data

Only run the next two cells if you're working with raw data. If you want to work with already trimmed and clean data, skip to the section 'Loading Clean Data'

The preprocess function mostly accomplishes the same things as Andrew's original data pipeline. To summarise:

    1) Convert's from microseconds to seconds
    2) Rearranges columns based on convention
    3) Applies the force divider function and converts the FSR values to Newtons

```python
# Force divider function
def force_divider(x):
    k_outer = 3.738
    k_inner = 0.368
    dist = x/k_inner
    return (k_outer + k_inner)*dist

# FSR model function
def FSR_model(x):
    return 131.2*np.exp(0.7801*x)/1000
```

```python
def preprocess(raw_data):
    raw_data = raw_data.drop(columns = ["IND"])
    raw_data["time"] = raw_data["time"].apply(lambda x: (x-raw_data["time"][0])*1e-6)
    raw_data = raw_data.set_index("time")
    raw_fsr_data = raw_data[["FSR_1", "FSR_2", "FSR_3", "FSR_4", "FSR_5", 
                        "FSR_6", "FSR_7", "FSR_8", "FSR_9", "FSR_10", "FSR_11", "FSR_12"]]
    raw_fsr_data = raw_fsr_data[["FSR_9", "FSR_10", "FSR_2", "FSR_1", "FSR_5", "FSR_6", 
                             "FSR_7", "FSR_8", "FSR_12", "FSR_11", "FSR_4", "FSR_3"]]
    raw_fsr_data.columns = ["FSR_1", "FSR_2", "FSR_3", "FSR_4", "FSR_5", 
                            "FSR_6", "FSR_7", "FSR_8", "FSR_9", "FSR_10", "FSR_11", "FSR_12"]
    fsr_data = raw_fsr_data[["FSR_1", "FSR_2", "FSR_3", "FSR_4", "FSR_5", 
                         "FSR_6", "FSR_7", "FSR_8", "FSR_9", "FSR_10", "FSR_11", "FSR_12"]] = raw_fsr_data[[
    "FSR_1", "FSR_2", "FSR_3", "FSR_4", "FSR_5", "FSR_6", "FSR_7", "FSR_8", "FSR_9", "FSR_10", "FSR_11", "FSR_12"]].apply(
    lambda x: force_divider(FSR_model(x*(5/1023))), axis=1).copy()
    return fsr_data
```

## Bootstrapping and Generating More Data

```python
def add_noise(x, sd = 0.5):
    for i in range(12):
        r = norm.rvs(size=2000, loc = 0, scale = sd)
        x.iloc[:,i] = x.iloc[:,i] + r
    return x

def compress(data):
    t = data.reset_index()
    t = t.groupby(t.index[:]//2.5).mean()
    t = t.set_index('time')
    return t
```

Set sample size and SD of random noise.

```python
size = 100
sd = 0.5

new_squeeze_data = [0]*size
sample = np.random.choice(np.arange(23), size)
for i in range(len(sample)):
    reshaped = squeeze_trimmed[sample[i]].set_index('time')
    com = compress(reshaped)
    new_squeeze_data[i] = add_noise(com,sd)
    
new_drop_data = [0]*size
sample = np.random.choice(np.arange(16), size)
for i in range(len(sample)):
    new_drop_data[i] = add_noise(drop_trimmed[sample[i]].set_index('time'), sd)
```

The next cell reshapes the data to output a 2D feature matrix (training data) and a 1D array of training labels.

## Modelling and Prediction

### K - Fold Cross Validation

```python
X = train_data
y = train_labels
skf = StratifiedKFold(n_splits=4)
```

```python
#Naive Bayes Implementation 
for train_index, test_index in skf.split(X, y):
    X_train, X_test, y_train, y_test = X[train_index], X[test_index], y[train_index], y[test_index]
    nb = GaussianNB().fit(X_train, y_train)
    print(nb.score(X_test, y_test))
```

```python
#Random Forest Implementation 
for train_index, test_index in skf.split(X, y):
    X_train, X_test, y_train, y_test = X[train_index], X[test_index], y[train_index], y[test_index]
    rf = RandomForestClassifier().fit(X_train, y_train)
    print(rf.score(X_test, y_test))
```

```python
#Decision Tree Implementation
for train_index, test_index in skf.split(X, y):
    X_train, X_test, y_train, y_test = X[train_index], X[test_index], y[train_index], y[test_index]
    dt = DecisionTreeClassifier().fit(X_train, y_train)
    print(dt.score(X_test, y_test))
```

## Feature Engineering/ EDA

```python
# Taken from Akhil's Features
#feature function block 
def impulse(data_matrix):
    #total impulse function for a single matrix of data 
    #iterates over data and computes the total impulse for all sensors over time 
    fsr_data = data_matrix[:,1:13]
    impulse = 0
    for col in np.arange(1,13):
        single_FSR = data_matrix[:,col]
        time = data_matrix[:,0]
        for i in np.arange(0,len(time)-1):
            delta_t = time[i+1]- time[i]
            averagef = (single_FSR[i+1] + single_FSR[i])/2 
            impulse = impulse + averagef*delta_t
    return impulse 


def jerk(data_matrix): 
    #iterates over data and computes the max jerk for each sensor over time
    #takes the max jerk over all FSR sensors and returns  
    fsr_data = data_matrix[:,1:13]
    jerk_array = []
    for col in np.arange(1,13):
        single_FSR = data_matrix[:,col]
        time = data_matrix[:,0]
        all_jerks = []
        for i in np.arange(0,len(time)-1):
            delta_t = time[i+1] - time[i]
            delta_f = single_FSR[i+1] - single_FSR[i]
            all_jerks.append(delta_f/delta_t)
        jerk_array.append(np.amax(all_jerks))
    max_jerk = np.amax(jerk_array)
    return max_jerk

def max_force(data_matrix):
    #max force readings across all sensors for a single matrix of data 
    fsr_data = data_matrix[:,1:13]
    force_array = [] 
    for col in np.arange(1,13):
        single_FSR = data_matrix[:,col]
        max_force_sensor = np.amax(single_FSR)
        force_array.append(max_force_sensor)
    max_force = np.amax(force_array)
    return max_force
```
